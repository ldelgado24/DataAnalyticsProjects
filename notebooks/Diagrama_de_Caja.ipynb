```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carga el archivo Excel
Homicidios = pd.read_excel(r'C:\Users\lubin\Documents\Proyectos\Entrenamiento\Diagrama.xlsx', 
                           sheet_name="Hoja1",  # Nombre de la hoja
                           header=0)  # Ajusta el encabezado según corresponda

# Ver los datos estadísticos de la variable 'Comunas' (si es numérica)
print(Homicidios['Comunas'].describe())



# Realizar el diagrama de caja para la variable 'Com'
plt.figure(figsize=(8, 6))
sns.boxplot(y='Comunas', data=Homicidios)
plt.title("Concentración de los Datos por Comunas")
plt.ylabel("Comunas")
plt.savefig("diagramadebigote.png")  # Especifica el nombre y formato de archivo
plt.show()
```
